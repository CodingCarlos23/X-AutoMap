```python
from bluesky import RunEngine
from bluesky.plans import scan
from bluesky.callbacks import LivePlot
from ophyd.sim import motor, det

RE = RunEngine({})

# Live plot the detector value as the motor moves
live_plot = LivePlot('det', 'motor')

# Perform a simple 1D scan of the motor while reading the detector
RE(scan([det], motor, 0, 10, 11), live_plot)
```
